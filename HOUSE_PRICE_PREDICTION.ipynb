```python
varlist=['mainroad','guestroom','hotwaterheating','airconditioning','prefarea']
```

```python
def binary_map(x):
    return x.map({'yes':1,'no':0})
data[varlist]=data[varlist].apply(binary_map)
```

```python
data.head()
```

```python
status=pd.get_dummies(data['furnishingstatus'])
```

```python
status.head()
```

```python
status=pd.get_dummies(data['furnishingstatus'],drop_first=True)
```

```python
status.head()
```

```python
data=pd.concat([data,status],axis=1)
```

```python
data.drop(['furnishingstatus'],axis=1,inplace=True)
```

```python
pd.DataFrame(data.corr().round(2))
```

```python
y=data['price']
x=data['area']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lgr=LinearRegression()
```

```python
x_train = np.array(x_train).reshape(-1, 1)
```

```python
lgr.fit(x_train,y_train)
```

```python
x_test=np.array(x_test).reshape(-1,1)
```

```python
predicted=lgr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test,predicted))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,predicted)
```
